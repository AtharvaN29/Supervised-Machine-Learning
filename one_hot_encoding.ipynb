```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df= pd.read_csv('./datasets_for_tuts/carprices.csv')
df
```

```python
dummies=pd.get_dummies(df.Car_Model)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car_Model','Audi A5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
X = final.drop(['Sell_Price'],axis='columns')
X
```

```python
y= final.Sell_Price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[10000,3,0,0]])
```

```python
model.score(X,y)
```
